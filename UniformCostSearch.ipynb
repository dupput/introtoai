```python
tree = {
    'S': {
        'A': 3,
        'B': 2,
    },
    'A': {
        'C': 3
    },
    'B': {
        'C': 6,
        'D': 9,
    },
    'C': {
        'E': 3
    },
    'D' : {
        'G': 3,
        'F': 8
    },
    'E': {
        'F': 2
    },
    'F': {
        'G': 2
    }
}
```

```python
paths = {}
solution = ['pathname', 10**8]
allNodes = list(tree.keys())
allNodes.remove('S')
start = 'S'
finish = 'G'

parentNode = 'S'
parentPath = 'S'
parentCost = 0

while True:
    queue = tree[parentNode].keys()

    for node in queue:
        pathname = parentPath + node
        cost = parentCost + tree[parentNode][node]

        paths[pathname] = cost
        
        if (node == finish) & (cost < solution[1]):
            solution = [pathname, cost]
        
    
    print('Current path tree: ', paths)
    minCost = 10**8
    for path, cost in paths.items():
        if cost < minCost:
            if path[-1] in allNodes:
                minCost = cost
                minPath = path
    
    parentPath = minPath
    parentNode = minPath[-1]
    parentCost = minCost
    
    
    print('Remaining unexplored nodes:', allNodes)
    if len(allNodes) == 0:
        break
    else:
        allNodes.remove(parentNode)
        del(paths[parentPath])
        print('Expanding path:', minPath)
        
print(solution)
```
