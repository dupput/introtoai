```python
df = pd.DataFrame({
    'Driver': [0,0,1,2,2,1,0,2,0],
    'engine': [1,2,1,1,0,2,2,2,1],
    'mod': [0,1,0,1,1,1,0,0,1],
    'prev': [1,1,0,0,1,0,0,1,1],
})
y = [1,1,0,1,0,1,0,0,0]
```

**Perceptron Algorythm**

```python
df = pd.DataFrame({'x1':[-1,-1,2,2,2,0], 'x2':[3,0,-5,-3,4,-2]})
X = df.values
y = [1,-1,-1,-1,-1,-1]
```

```python
clf = Perceptron(max_iter=2, tol=0.0001, random_state=0)
clf.fit(df, y)
predictions = clf.predict(df)

test_acc_perceptron = accuracy_score(y, predictions)
print('The accuracy of the perceptron classifier on the test set is %.2f' % test_acc_perceptron)
```

**Decision Tree Classifier**

```python
clf = DecisionTreeClassifier(random_state=0, criterion='entropy', max_depth=3)
clf.fit(df, y)
predictions = clf.predict(df)

accuracy = accuracy_score(y, predictions)
print('The accuracy of the decision tree classifier on the test set is %.2f' % accuracy)
```
