```python
tree = {
    'S': {
        'C': {
            'g(n)' : 5,
            'h(n)' : 5
        },
        'B': {
            'g(n)' : 1,
            'h(n)' : 3
        }
    },
    'A': {
        'S': {
            'g(n)' : 7,
            'h(n)' : 5
        }
    },
    'B':{
        'E': {
            'g(n)' : 2,
            'h(n)' : 3
        },
        'F': {
            'g(n)' : 4,
            'h(n)' : 1
        }
    },
    'D': {
        'A': {
            'g(n)' : 2,
            'h(n)' : 9
        },
        'G2': {
            'g(n)' : 2,
            'h(n)' : 0
        }
    },
    'E': {
        'D': {
            'g(n)' : 1,
            'h(n)' : 2
        },
        'G2': {
            'g(n)' : 4,
            'h(n)' : 0
        }
    },
    'F': {
        'G1': {
            'g(n)' : 2,
            'h(n)' : 0
        }
    }
}
```

```python
frontier = set('S')
explored = []
paths = {}

start = 'S'
finish = ['G1', 'G2']

currentNode = start
parentCost = 0
parentPath = 'S'

while True:
    print(currentNode)
    queue = list(tree[currentNode].keys())
    print(queue)
    explored.append(currentNode)
    
    for node in queue:
        if node in explored:
            continue
        
        frontier.add(node)
        
        path = parentPath + node
        
        cost = tree[currentNode][node]['g(n)'] + parentCost
        hCost = tree[currentNode][node]['h(n)'] + cost
        
        paths[path] = {
            'cost': cost,
            'hcost': hCost,
            'endNode': node
        }
    frontier.remove(currentNode)
    
    print('Explored: ', explored)
    print('Frontier: ', frontier)
    print('Paths: ', paths)
    
    max_hcost = 10**8
    for path, data in paths.items():
        node = data['endNode']
        if node in explored:
            continue
        
        cost = data['cost']
        hcost = data['hcost']
            
        if hcost < max_hcost:
            max_hcost = hcost
            parentPath = path
            parentCost = cost
            currentNode = node
    
    print('parentPath: ', parentPath)
    del paths[parentPath]
    
    if currentNode in finish:
        print(currentNode, 'has been reached')
        break
    


print('Optimal route: ', parentPath)
```
