```python
tree = {
    'S': {
        'A': 3,
        'B': 1,
        'C': 6,
        'D': 3
    },
    'A': {
        'E': 7,
        'G1': 10
    },
    'B': {
        'C':3,
        'F':2
    },
    'C': {
        'G3': 11
    },
    'D': {
        'B': 5,
        'G2': 6
    },
    'E': {
        'G1': 1
    },
    'F': {
        'D': 3
    }
}
```

```python
explored = set()
frontier = ['S']

paths = {}
start = 'S'
finish = ['G1', 'G2', 'G3']

currentNode = start
parentCost = 0
parentPath = 'S'

while True:
    print(currentNode)
    queue = list(tree[currentNode].keys())
    print(queue)
    explored.add(currentNode)
    
    for node in queue:
        if node in explored:
            continue
        
        frontier.append(node)
        
        path = parentPath + node
        
        cost = tree[currentNode][node] + parentCost
        
        paths[path] = {
            'cost': cost,
            'endNode': node
        }
    frontier.remove(currentNode)
    
    print('Explored: ', explored)
    print('Frontier: ', frontier)
    print('Paths: ', paths)
    
    max_cost = 10**8
    for path, data in paths.items():
        node = data['endNode']
        if node in explored:
            continue
        
        cost = data['cost']
            
        if cost < max_cost:
            max_cost = cost
            parentPath = path
            parentCost = cost
            currentNode = node
    
    print('parentPath: ', parentPath)
    del paths[parentPath]
    
    if currentNode in finish:
        print(currentNode, 'has been reached')
        break

print('Optimal route: ', parentPath)
```
